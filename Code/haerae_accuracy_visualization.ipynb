```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
index = ["introduce_typo", "drop_jongsung", "repeat_char", "merge_words", "swap_parts"]

haerae_1_0_gk_EXAONE_data = {
    "level1": [36.36, 36.36, 36.93, 36.36, 34.09],
    "level2": [31.25, 32.95, 34.66, 35.80, 32.39],
    "level3": [31.82, 32.39, 33.52, 36.36, 34.09],
    "level4": [28.98, 31.82, 34.09, 35.23, 32.39],
    "level5": [29.55, 32.39, 32.95, 36.36, 34.66]
}
df = pd.DataFrame(haerae_1_0_gk_EXAONE_data, index=index)

plt.figure(figsize=(10, 6))
for typo in df.index:
    plt.plot(df.columns, df.loc[typo], marker='o', label=typo)

plt.axhline(y=35.23, color='gray', linestyle='--', label='No Typo Baseline (35.23)')

plt.title("Accuracy per Typo Type across Levels\n(HAERAE_1.0_GK, Model: EXAONE-3.5-2.4B-Instruct)")
plt.xlabel("Typo Level")
plt.ylabel("Accuracy (%)")
plt.ylim(25, 39)
plt.grid(True)
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```
